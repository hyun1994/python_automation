```python
import pandas as pd
from googleapiclient.discovery import build
```

```python
api_obj = build('youtube', 'v3', developerKey="AIzaSyAE3iD819cda3UeCBd8lTA_15TmpdglwZs")
response = api_obj.commentThreads().list(part='snippet, replies', videoId='x-zRcLkQ2fg', maxResults=100).execute()
```

```python
comments = list()

while response:
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        comments.append([comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])

        if item['snippet']['totalReplyCount'] > 0:
            for reply_item in item['replies']['comments']:
                reply = reply_item['snippet']
                comments.append([reply['textDisplay'], reply['authorDisplayName'], reply['publishedAt'], reply['likeCount']])

    if 'nextPageToken' in response:
        response = api_obj.commentThreads().list(part='snippet, replies', videoId = 'x-zRcLkQ2fg', pageToken=response['nextPageToken'], maxResults=100).execute()
    else:
        break
```

```python
df = pd.DataFrame(comments, columns = ['comment', 'author', 'at', 'like_cnt'])
```
