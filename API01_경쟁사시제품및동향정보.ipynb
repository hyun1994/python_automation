```python
import pandas as pd
df = pd.DataFrame(result['items'])
```

```python
import re

def del_tag(s):
    s = re.sub('(<([^>]+)>)', '', s)
    s = s.replace('&quot;', '')
    return s

df['title'] = df['title'].apply(del_tag)
df['description'] = df['description'].apply(del_tag)
df
```

```python
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
twitter.pos(df["title"][0], norm=True, stem=True)
```

```python
temp = twitter.pos(df["title"][0], norm=True, stem=True)
pos_list = [t[0] for t in temp if t[1] == "Noun"]
```

```python
pos_list
```

```python
def to_pos(text):
    temp = twitter.pos(text, norm=True, stem=True)
    pos_list = [t[0] for t in temp if t[1] == "Noun"]
    return pos_list
```

```python
tot_list = list()
for t in df["title"]:
    tot_list = tot_list + to_pos(t)
```

```python
import collections
tot_dict = collections.Counter(tot_list)
```
