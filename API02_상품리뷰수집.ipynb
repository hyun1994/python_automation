```python
def to_eng(text):
    """
    영어로 번역하는 함수
    """

    import requests

    try:
        #네이버 파파고 api 이용하기
        request_url = "https://openapi.naver.com/v1/papago/n2mt"
        headers = {"X-Naver-Client-Id": "mdf3sxzPDGQ_E5Z7NES7", "X-Naver-Client-Secret": "0hZUfD7drC"}
        params = {"source":"ko", "target" : "en", "text" : text}
        response = requests.post(request_url, headers=headers, data=params)
        result = response.json()

        return result["message"]['result']["translatedText"]
    except Exception as e:
        print(e)
        pass
    return ""
```

```python
df['eng'] = df['description'].apply(to_eng)
df
```

## 3. 감성분석

```python
# 감성분석 해보기
from textblob import TextBlob

analysis = TextBlob(df['eng'][0])
posneg = analysis.sentiment.polarity
```

```python
print(df['description'][0])
print(df['eng'][0])
print(posneg)

if posneg == 0:
    print("NEUTRAL")
if posneg > 0:
    print("POSITIVE")
elif posneg < 0:
    print("NEGATIVE")
```

```python
def to_posneg(text):
    analysis = TextBlob(text)
    posneg = analysis.sentiment.polarity
    return posneg
```

```python
df['posneg'] = df['eng'].apply(to_posneg)
```
