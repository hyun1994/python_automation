# 2. 함수만들기

```python
def get_div(ticker, from_day, to_day):

    url=f'https://api.tiingo.com/tiingo/daily/{ticker}/prices?startDate={from_day}&endDate={to_day}&token={token}'
    response = requests.get(url)
    result = response.json()
    df = pd.DataFrame(result)

    div = df['divCash'].sum()
    current = df.tail(1)['close'].values[0]

    return [ticker, div, current]
```

# 4. S&P 500 종목의 주가와 배당금, 배당수익률 구하기

```python
from tqdm import tqdm
```

```python
sp500.head()
```

```python
tot_list = list()

for s in tqdm(sp500['Symbol']):
    try:
        tot_list.append(get_div(s, from_day, to_day))
    except Exception as e:
        print(f'{s} has error. {e}')
```

```python
df_div = pd.DataFrame(data=tot_list, columns=['ticker', 'div', 'current'])
```

```python
df_div['div_rate'] = df_div['div']/df_div['current'] * 100
```

```python
df_div.head()
```

```python
df_div = df_div.sort_values('div_rate', ascending=False)
```

```python
def get_oi(ticker):
    df_yf = yf.Ticker(ticker)
    oi_mean = df_yf.financials.T['Operating Income'].mean()
    oi = df_yf.financials.T['Operating Income'].head(1).values[0]
    oi_rate = oi/oi_mean

    return [ticker, oi_rate]
```

```python
import time
import random

def random_sleep(fr=5, to=10):
    time.sleep(random.randint(fr,to)/10)
```

```python
tot_list = list()

for s in tqdm(df_div['ticker']):
    try:
        tot_list.append(get_oi(s))
    except Exception as e:
        print(f'{ticker} has error. {e}')
```

```python
df_oi = pd.DataFrame(data=tot_list, columns=['ticker', 'oi_rate'])
```

```python
df = pd.merge(df_div, df_oi, how='left', on='ticker')
```
